```python
from statsmodels.tsa.stattools import adfuller
```

```python
deaths1 = deaths.loc[:,'Lat':'3/23/20']
```

```python
print(" > Is the data stationary ?")
for i in deaths1.columns:
    dftest = adfuller(deaths1[i], autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("Critical values :")
    for k, v in dftest[4].items():
        print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
```

```python
recovered1 = recovered.loc[:,'Lat':'3/23/20']
```

```python
print(" > Is the data stationary ?")
for i in recovered1.columns:
    dftest = adfuller(recovered1[i], autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("Critical values :")
    for k, v in dftest[4].items():
        print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
```

```python
confirmed1 = confirmed.loc[:,'Lat':'3/23/20']
```

```python
print(" > Is the data stationary ?")
for i in confirmed1.columns:
    dftest = adfuller(confirmed1[i], autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("Critical values :")
    for k, v in dftest[4].items():
        print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
```
